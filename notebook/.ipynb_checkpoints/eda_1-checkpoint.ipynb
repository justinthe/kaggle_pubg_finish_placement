```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
datadir = '../data/raw'
train = pd.read_csv(os.path.join(datadir, 'train_V2.csv'))
test = pd.read_csv(os.path.join(datadir, 'test_V2.csv'))
```

```python
print("Shape: ",train.shape)
print("-"*20)
print("Cols:", train.columns)
print("-"*60)
train.head()
```

```python
plt.figure(figsize=(8,5))
train['winPlacePerc'].plot.hist(rwidth=0.9)
```

```python
# Check null val
train.isnull().sum()
```

```python
train[train['winPlacePerc'].isnull()]
```

```python
# null in train most likely means 0 
train = train.fillna(0)
```

```python
# check correlation
num_data_corr = train.select_dtypes(exclude=(['object'])).corr()
```

```python
mask = np.zeros_like(num_data_corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(num_data_corr, mask=mask)
```

```python
corr_list = num_data_corr['winPlacePerc'].sort_values(ascending=False)
corr_list[1:11]
```

Seems that moving is the KEY. By Moving, player get to acquire more boost, weapon, meet more enemy -thus get more kill-.

```python
plt.figure(figsize=(18, 8))
# check the first 6
for i in range(6):
    ii = '23' + str(i+1)
    plt.subplot(ii)
    feat = corr_list.index.values[i+1]
    plt.scatter(train[feat], train['winPlacePerc'])
    sns.regplot(x=feat, y='winPlacePerc', data=train, scatter=False, color='red')
    plt.title(feat)
```

```python
# same group of people playing how many times
train.groupby(['groupId'], as_index=False)['matchId'].agg('count')
```

```python
# number of group per matches
group_per_match = train.groupby(['matchId'], as_index=False)['groupId'].agg('count')
group_per_match.columns = ['matchId','group_cnt']
group_per_match
```

```python
train = pd.merge(train, group_per_match, how='left', on='matchId')
train.head(10)
```

```python
plt.figure(figsize=(18, 8))
train['group_cnt'].value_counts().head(15).plot.bar()
plt.xlabel('group count in matches')
plt.title('top 15 of group count in matches')
```

```python
# Combine train and test data
print("Train: ",train.shape)
print("-"*20)
print("Test:", test.shape)
print("-"*20)
print("Train Columns: ",train.columns)
print("-"*20)
print("Test Columns:", test.columns)
```
